```r
source("../../BrusselSprouts/scripts/functions.R")
```

```r
scripts_dir="/home/jnrunge/data/trd/scripts/"
initial_timedate=Sys.time()
```

```r
samples=readLines("~/data/trd/full2489Matrix.vcf.gz.samples")
head(samples)
```

```r
for(s in samples){
    if(file.exists(paste0("~/data/TRD/Oppo-Homo-Pos/",s,".loci.gz"))){
        next
    }
    cmd=paste0("bcftools view -V indels,mnps,other -s ",s," ~/data/trd/full2489Matrix.vcf.gz | bcftools view -i 'F_MISSING==0 & GT[0] == \"hom\" & GQ >=30 && FMT/DP >= 20' | bcftools view -e 'type!=\"snp\"' | grep -v ^# | cut -f 1,2 | gzip > ~/data/TRD/Oppo-Homo-Pos/",s,".loci.gz")
    #print(cmd)
    sbatch_list=execute_complex_sbatch(cmd, mem="4G", cores="1", env="bwaetc", jobname = "OHLOCI",jobs_simul = 10,uniqueRunID = s, scripts_dir = scripts_dir, jobs_total = 10,time="long", initial_timedate = initial_timedate)
    
}
```

```r
if(exists("sbatch_list")){
    start_sbatch_list(sbatch_list, 10, "OHLOCI", initial_timedate)
    }
```
