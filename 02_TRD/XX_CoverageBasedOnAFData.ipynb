```r
library(tidytable)
source("~/BrusselSprouts/scripts/functions.R")

files=list.files(path = "/home/jnrunge/data/TRD/results/shiny", pattern="AF.csv.gz$", full.names = TRUE)
```

```r
getDepthFromAFfile<-function(x){
    AF<-fread(x)
    return(summarize(AF, meanDepth=mean(sumCount))%>%
          mutate(file=basename(x)))
}

AF_depths<-bind_rows(lapply(files, getDepthFromAFfile))
```
