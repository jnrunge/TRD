```r
library(vcfR)
```

```r
files=system(command="ls /home/jnrunge/data/TRD/long_read_Andreas/*.vcf", intern=TRUE)
```

```r
SV_df=data.frame(file=files, SVLEN5k = NA, SVLEN10k = NA, SVLEN100k=NA, stringsAsFactors = FALSE)
```

```r
for(i in 1:nrow(SV_df)){
    vcf<-read.vcfR(file = SV_df$file[i])

    info<-vcfR::extract_info_tidy(vcf)

    print(ggplot(mapping=aes(as.numeric(info$SVLEN)))+
    geom_histogram(binwidth = 1000)+
    geom_vline(xintercept = c(-5000,5000), color="red"))
    
    SV_types_tmp=summarise(group_by(info, SVTYPE),n=n())
    SV_types_tmp$file=SV_df$file[i]
    if(i==1){
        SV_types=SV_types_tmp
    }else{
        SV_types=bind_rows(SV_types,SV_types_tmp)
    }

    SV_df$SVLEN5k[i]=sum(as.numeric(info$SVLEN) > 5000 | as.numeric(info$SVLEN) < -5000)
    SV_df$SVLEN10k[i]=sum(as.numeric(info$SVLEN) > 10000 | as.numeric(info$SVLEN) < -10000)
    SV_df$SVLEN100k[i]=sum(as.numeric(info$SVLEN) > 100000 | as.numeric(info$SVLEN) < -100000)

    }
```

```r
# manual lookup for now based on 1002_strains_total_set$final_1002_site & Collection_YJS_Box
lookup=data.frame(row.names=c("YJS7207","YJS7234","YJS7273","YJS7281","YJS7284"),
             val=c(NA,"BET","BAN","CMQ",NA))
lookup
```

```r
samples_good=unlist(lapply(unlist(lapply(unlist(lapply(files_good,getLast,split="/")),getFirst_v2,split="MUM")),getLast,split="_"))
for(i in 1:length(samples_good)){
    s=samples_good[i]
    if(s %in% rownames(lookup)){
        samples_good[i]=lookup$val[which(rownames(lookup)==s)]
    }
}
samples_good=samples_good[!is.na(samples_good)]
samples_good
```
